```python
import pandas as pd
import datetime as dt

data = pd.read_excel("Orders-and-Returns-Sample-Superstore-Data-Workbook.xlsx", sheet_name = "Orders")
```

```python
data['Order_Month'] = data['Order Date'].dt.month
```

```python
data.groupby(["State or Province","Order_Month"])[["Profit", "Quantity ordered new"]].agg("sum")
```
